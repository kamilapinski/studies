```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

## Training a classifier [do not modify]

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

## GradCAM implementation (add your code)

```python
class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
        grad_cam = GradCAM(model=model, target_layers=[layer1, layer2])
        grayscale_cam = grad_cam(input_tensor, targets=[class_id])
    """

    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
        self.model = model
        self.target_layers = target_layers

    def __call__(self, input_tensor: Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        """
        Returns: numpy array of shape (B, H, W), values 0..1.
        """
        # we have to force calculating gradients
        with torch.enable_grad():
            is_training = self.model.training
            self.model.eval() # model cannot be trained here

            self.activations = {}
            self.gradients = {}

            def save_activation(module, input, output):
              self.activations[module] = output

            def save_gradient(module, grad_in, grad_out):
              # one output of the model, so grad_out[0]
              self.gradients[module] = grad_out[0]

            hooks = []
            try:
              # register hooks
              for target_layer in self.target_layers:
                hooks.append(target_layer.register_forward_hook(save_activation))
                hooks.append(target_layer.register_full_backward_hook(save_gradient))

              input_tensor = input_tensor.to(device)
              input_tensor.requires_grad = True
              B, C, H, W = input_tensor.shape

              # forward pass
              output = self.model(input_tensor)

              if targets is None:
                # when we don't provide targets we take the class with the highest probability
                target_class_indices = output.argmax(dim=-1) # (B,)
              else:
                target_class_indices = torch.tensor(targets, device=input_tensor.device) # (B,)

              target_score = torch.stack([output[i, target_class_indices[i]] for i in range(B)])

              # backward
              self.model.zero_grad()
              # we run backward once for the entire batch (B). The .sum() operation is used to create a single scalar target,
              # which enables the simultaneous calculation of gradients for all B elements in the batch.
              # retain_graph=True is necessary to keep the computation graph intact for subsequent analysis of other target layers.
              target_score.sum().backward(retain_graph=True)

              cams = []
              for target_layer in self.target_layers:
                A_k = self.activations[target_layer]
                grad_yc_Ak = self.gradients[target_layer]
                alpha_k_c = grad_yc_Ak.mean(dim=(-1, -2), keepdim=True)

                L_c = (alpha_k_c * A_k).sum(dim=1, keepdim=True)
                L_c = torch.relu(L_c)

                cam = F.interpolate(L_c, size=(H, W), mode='bilinear', align_corners=False)
                # normalize to 0..1
                cam = cam.squeeze(1) # shape (B, H, W)
                cam_min = cam.amin(dim=(-1, -2), keepdim=True)
                cam_max = cam.amax(dim=(-1, -2), keepdim=True)
                # avoid division by zero if cam_max == cam_min
                cam = (cam - cam_min) / (cam_max - cam_min + 1e-7)

                cams.append(cam.detach().cpu().numpy())
            finally:
              for hook in hooks:
                hook.remove()

              if is_training:
                self.model.train()

            if not cams:
              return np.array([])

            stacked_cams = np.stack(cams, axis=0) # Shape: (Num_Layers, B, H, W)
            averaged_cam = np.mean(stacked_cams, axis=0) # Shape: (B, H, W)

            return averaged_cam
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

### get_approximate_mask

```python
model.eval()
global_grad_cam = GradCAM(model=model, target_layers=target_layers)

def get_approximate_mask(image_tensor: Tensor, H: int, W: int) -> np.ndarray:
    grayscale_cam = global_grad_cam(image_tensor, targets=None)
    heatmaps_np = grayscale_cam # (H, W), values 0..1

    # binarize the heatmap using threshold
    THRESHOLD_VALUE = 0.5

    binary_mask = heatmaps_np >= THRESHOLD_VALUE
    approx_mask_np = (binary_mask * 255).astype(np.uint8)

    final_masks = approx_mask_np

    return final_masks
```

### SamPipelineEvaluator class and metric functions

```python
def calculate_center_of_mass(mask: np.ndarray) -> tuple[float, float]:
    """Calculates the (x, y) centroid of a binary mask."""
    # note: mask should have foreground pixels as 1 or 255
    moments = cv2.moments(mask.astype(np.uint8))

    # m00 - sum of all pixels in the mask
    # m10 - sum(x*I)
    # m01 - sum(y*I)
    if moments["m00"] == 0:
        # fallback to center of the image if mask is empty
        H, W = mask.shape
        return W / 2, H / 2

    # cx = m10 / m00, cy = m01 / m00
    cx = moments["m10"] / moments["m00"] # x of centroid
    cy = moments["m01"] / moments["m00"] # y of centroid
    return cx, cy

def calculate_iou(pred_mask: np.ndarray, gt_mask: np.ndarray) -> float:
    intersection = np.logical_and(pred_mask, gt_mask).sum(axis=(1, 2))
    union = np.logical_or(pred_mask, gt_mask).sum(axis=(1, 2))

    iou = np.zeros_like(union, dtype=float)

    valid_mask = union > 0
    iou[valid_mask] = intersection[valid_mask] / union[valid_mask]
    empty_mask = union == 0
    iou[empty_mask] = 1.0

    return iou

def euclidean_distance(p1: np.ndarray, p2: np.ndarray) -> float:
    # (B, 2)
    return np.sqrt(np.sum((p1 - p2) ** 2, axis=1))

def furthest_corner(center_point: np.ndarray, H: int, W: int):
    A = np.array([5, 5])
    B = np.array([5, H - 6])
    C = np.array([W - 6, H - 6])
    D = np.array([W - 6, 5])

    corners = [A, B, C, D]
    distances = [euclidean_distance(np.expand_dims(center_point, axis=0), np.expand_dims(corner, axis=0))[0] for corner in corners]
    return corners[np.argmin(distances)]

class SamPipelineEvaluator:
    def __init__(self, pipeline: BasicSamPipeline, dataset_loader):
        self.pipeline = pipeline
        self.dataset_loader = dataset_loader

    def evaluate(self) -> dict:
        """Runs the pipeline over the dataset and computes all metrics."""

        all_iou = []
        all_hit_rates = []
        all_distances = []
        hard_images = []
        hard_predicted_masks = []
        hard_gt_masks = []
        progress_bar = tqdm(self.dataset_loader, desc="")

        for images, results, gt_masks in progress_bar:
            if isinstance(self.pipeline, ForegroundOnlySamPipeline) or isinstance(self.pipeline, ForegroundBackgroundSamPipeline):
                point_coords, point_labels = self.pipeline.generate_points(images)
                fg_points_generated = point_coords[:][point_labels == 1]
            else:
                B, H, W = gt_masks.shape
                # BasicSamPipeline (center of image)
                fg_points_generated = np.array([[W // 2, H // 2] for _ in range(B)])

            y_coords, x_coords = fg_points_generated[:, 1], fg_points_generated[:, 0]
            is_hit_np = gt_masks.numpy()[:, y_coords, x_coords]
            all_hit_rates.append(is_hit_np)

            with torch.no_grad():
                predicted_masks_tensor = self.pipeline(images) # (B, H, W)

            predicted_masks_np = predicted_masks_tensor.numpy()
            ious = calculate_iou(predicted_masks_np, gt_masks.numpy())
            all_iou.append(ious)

            low_iou_indices = np.where(ious < 0.5)[0]

            if len(low_iou_indices) > 0:
              hard_images.append(images.numpy()[low_iou_indices])
              hard_predicted_masks.append(predicted_masks_np[low_iou_indices])
              hard_gt_masks.append(gt_masks.numpy()[low_iou_indices])

            gt_centers = []
            for j in range(len(images)):
              image_tensor, _, gt_mask_tensor = images[j].cpu(), results[j], gt_masks[j].cpu()
              gt_mask_np = gt_mask_tensor.numpy()
              # distance
              gt_center_x, gt_center_y = calculate_center_of_mass(gt_mask_np)
              gt_center = np.array([gt_center_x, gt_center_y])
              gt_centers.append(gt_center)

            gt_centers = np.array(gt_centers)
            distances = euclidean_distance(fg_points_generated, gt_centers)
            all_distances.append(distances)

            progress_bar.set_postfix(
                distance=f"{np.mean(all_distances):.3f}",
                hit_rate=f"{np.mean(all_hit_rates):.1%}",
                iou=f"{np.mean(all_iou):.3f}"
            )


        # averages
        avg_iou = np.mean(all_iou)
        avg_hit_rate = np.mean(all_hit_rates)
        avg_distance = np.mean(all_distances)

        hard_images = np.concatenate(hard_images, axis=0)
        hard_predicted_masks = np.concatenate(hard_predicted_masks, axis=0)
        hard_gt_masks = np.concatenate(hard_gt_masks, axis=0)

        return {
            "Pipeline Class": self.pipeline.__class__.__name__,
            "IoU (Avg)": avg_iou,
            "Hit Rate (Avg)": avg_hit_rate,
            "Distance to GT Center (Avg)": avg_distance,
            "Hard samples": (hard_images, hard_predicted_masks, hard_gt_masks)
        }
```

### ForegroundOnlySamPipeline

```python
class ForegroundOnlySamPipeline(BasicSamPipeline):
    def generate_points(self, images: Tensor):
      B, C, H, W = images.shape
      point_coords = []
      point_labels = []

      approx_masks_np = get_approximate_mask(images, H, W)

      for i, image in enumerate(images):

          # center of mass
          moments = cv2.moments(approx_masks_np[i])
          if moments["m00"] == 0:
              # Fallback: use the image center
              cx, cy = W // 2, H // 2
          else:
              cx = int(moments["m10"] / moments["m00"])
              cy = int(moments["m01"] / moments["m00"])

          point_coords.append([(cx, cy)])
          point_labels.append([1])

      point_coords = np.array(point_coords, dtype=np.int64)
      point_labels = np.array(point_labels, dtype=np.int64)

      return point_coords, point_labels


    def __call__(self, images: Tensor) -> Tensor:
        point_coords, point_labels = self.generate_points(images)
        return self.segment(
            images, np.array(point_coords, dtype=np.int64), np.array(point_labels, dtype=np.int64)
        )


def example_foreground_sam():
    indices = [3, 10, 42]
    images, labels, gt_masks = next(iter(test_loader))
    images, labels, gt_masks = images[indices], labels[indices], gt_masks[indices]

    pipeline = ForegroundOnlySamPipeline()

    results = pipeline(images).cpu()

    for image, result, gt_mask in zip(images, results, gt_masks, strict=True):
        show_image_row(
            {
                "Input image": inverse_transform(image),
                "Basic SAM predicted mask": result,
                "Ground-truth mask": gt_mask,
            }
        )


example_foreground_sam()
```

### ForegroundBackgroundSamPipeline

```python
class ForegroundBackgroundSamPipeline(BasicSamPipeline):
    def generate_points(self, images: Tensor):
        B, C, H, W = images.shape
        point_coords = []
        point_labels = []

        approx_masks_np = get_approximate_mask(images, H, W)

        for i, image in enumerate(images):

            # center of mass
            moments = cv2.moments(approx_masks_np[i])
            if moments["m00"] == 0:
                # Fallback: use the image center
                cx, cy = W // 2, H // 2
            else:
                cx = int(moments["m10"] / moments["m00"])
                cy = int(moments["m01"] / moments["m00"])

            curr_point_coords = [(cx, cy), furthest_corner(np.array([cx, cy]), H, W)]
            curr_point_labels = [1, 0]

            point_coords.append(curr_point_coords)
            point_labels.append(curr_point_labels)

        point_coords = np.array(point_coords, dtype=np.int64)
        point_labels = np.array(point_labels, dtype=np.int64)

        return point_coords, point_labels

    def __call__(self, images: Tensor) -> Tensor:
        point_coords, point_labels = self.generate_points(images)

        return self.segment(
            images, np.array(point_coords, dtype=np.int64), np.array(point_labels, dtype=np.int64)
        )


def example_foreground_background_sam():
    indices = [3, 10, 42]
    images, labels, gt_masks = next(iter(test_loader))
    images, labels, gt_masks = images[indices], labels[indices], gt_masks[indices]

    pipeline = ForegroundBackgroundSamPipeline()

    results = pipeline(images).cpu()

    for image, result, gt_mask in zip(images, results, gt_masks, strict=True):
        show_image_row(
            {
                "Input image": inverse_transform(image),
                "Basic SAM predicted mask": result,
                "Ground-truth mask": gt_mask,
            }
        )


example_foreground_background_sam()
```
